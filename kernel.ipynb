```python
# this shows the average sale in 2017
average_sal=data["Sales"].mean()
print("Average_sale" ,average_sal)
```

```python
sales =data["Sales"].tolist()
temperature =data["Temperature"].tolist()
rainfall =data["Rainfall"].tolist()
relation ={'Sales' : sales  ,'Temperature' : temperature ,'Rainfall ' :rainfall}
df=pd.DataFrame(relation)
df.plot(style =['o' ,'rx'])
```

**This above  scatter plot graph which shows Sales and Temperature. We can see  that Sales increase as Temperature increases.**
